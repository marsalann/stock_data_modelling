## Feed Forward

Perform prediction by using feed-forward neural network (sliding window)

```python
x = np.array(df)[:,1:6]
max = np.max(x,axis=0)
minn = np.min(x,axis=0)

normalized_x = (x - min)/(max - min)

features = ['Open', 'High', 'Low', 'Close', 'Volume']
df_1 = pd.DataFrame(normalized_x, columns=features)
target_columns = features

# Define window size
window_size = 10

# Initialize empty lists to store models and histories
models = []
histories = []

# Iterate over each target column
for target_column in target_columns:
    # Prepare data for the target column
    sequences = []
    labels = []

    for i in range(len(df_1) - window_size):
        sequences.append(df_1[features].iloc[i:i+window_size].values)
        labels.append(df_1[target_column].iloc[i+window_size])

    X = np.array(sequences, dtype=np.float32)
    y = np.array(labels, dtype=np.float32)

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

    # Define and compile the model
    model = tf.keras.Sequential([
      tf.keras.layers.Flatten(input_shape=(window_size, len(features))),
      tf.keras.layers.Dense(64, activation='relu'),
      tf.keras.layers.Dense(1)  # Output layer with 1 neuron for regression
  ])

    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train the model
    history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split = 0.2)

    # Append the trained model and its history to the lists
    models.append(model)
    histories.append(history)

    # De-normalization

    # Make predictions on the test set
    predictions = model.predict(X_test)

    # De-normalization
    predictions_denormalized = predictions * (max - min) + min
    y_test_denormalized = y_test * (max - min)[target_columns.index(target_column)] + min[target_columns.index(target_column)]

    # Plot actual vs. predicted values
    plt.figure(figsize=(10, 5))
    plt.plot(y_test, label='Actual')
    plt.plot(predictions, label='Predicted')
    plt.title(f'{target_column} Prediction')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.legend()
    plt.show()
```
